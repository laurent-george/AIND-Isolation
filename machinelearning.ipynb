```python
import pandas

def gen_features():
    game = Board(agentUdacity.player, agentUdacity.player)
    winner, _, termination = game.play()
    res = 1 if winner == agentRandom1.player else -1
    vect = pandas.DataFrame()
    final_features = []
    for feature_vect in game.features:
        feature_vect["res"] = res #if feature_vect["move_count"] %2 == 0 else -(res)
        feature_vect["legal_move_square"] = feature_vect["player_legal_moves"] ** 2
        feature_vect["opp_move_square"] = feature_vect["opponent_legal_moves"] ** 2
       # if feature_vect["move_count"] %2 == 0:
        final_features.append(feature_vect)
    return final_features
```

```python
games = [gen_features() for x in range(100)]
```
